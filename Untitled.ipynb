```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor, GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, recall_score, precision_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import cross_val_score
```

```python
def mse(y_pred,y_true):
    yp = np.array(y_pred)
    yt = np.array(y_true)
    mse = np.mean((yp-yt)**2)
    return mse
```

```python
data_new = data.drop(['floor'],axis=1).copy()
```

```python
def cat_to_num(data_x,col):
    dummy = pd.get_dummies(data_x[col])
    del dummy[dummy.columns[-1]]#To avoid dummy variable trap
    data_x= pd.concat([data_x,dummy],axis =1)
    return data_x
```

```python
categorical = []
numerical = []
for column in data_new.columns:
    if data[column].dtype == object:
        categorical.append(column)
    else:
        numerical.append(column)
print('Categorical features: {}'.format(categorical)) 
print('Numerical features: {}'.format(numerical))  
data_new = cat_to_num(data_new,categorical)
print(data_new.columns)
data_new.head()
```

```python
data_new.info()
```

```python
data_x = data_new.drop(['city','animal','furniture','animal_not acept','hoa (R$)','area','property tax (R$)'],axis=1)
data_x.info()
```

```python
y = data_x['rent amount (R$)']
x = data_x.drop(['rent amount (R$)'],axis=1)
```

## Regression

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size =0.2,random_state = 42)
```

```python
regressors = {'LinearRegressor': LinearRegression(),
             'RandomForestRegressor': RandomForestRegressor(),
             'GradientBoostingRegressor': GradientBoostingRegressor()}
for keys,regressor in regressors.items():
    regressor.fit(X_train,y_train)
    training_score = cross_val_score(regressor,X_train,y = y_train, scoring='r2',cv=5)
    print("Regressor: ", regressor.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
R2_score = cross_val_score(rfr,X_train,y_train,cv=5)
mse_score = cross_val_score(rfr,X_train,y_train,scoring = 'neg_mean_squared_error',cv=5)
```

```python
print('RandomForestRegressor R2: {}'.format(R2_score.mean()))
print('RandomForestRegressor MSE: {}'.format(mse_score.mean()))
```

```python
y_pred_rfr = rfr.predict(X_test)
r2 = r2_score(y_test,y_pred_rfr)
print('RandomForestRegressor R2: {}'.format(r2))
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(X_train,y_train)
```

```python
y_pred_gbr = gbr.predict(X_test)
r2 = r2_score(y_test,y_pred_gbr)
print('RandomForestRegressor R2: {}'.format(r2))
```
